```python
import pandas as pd
import pylab as plt
plt.style.use('fivethirtyeight')

def check_substring_in_file(p, y):
    """
    Check if a substring y exists in a file specified by a path p.

    Args:
    p (str): Path to the file.
    y (str): Substring to search for.

    Returns:
    bool: True if the substring is found, False otherwise.
    """
    try:
        # Open the file and read its contents
        with open(p, 'r', encoding='utf-8') as file:
            content = file.read()

            # Check if the substring exists in the content
            return y in content
    except FileNotFoundError:
        print("The file was not found.")
        return False
    except Exception as e:
        print(f"An error occurred: {e}")
        return False

def getsubtype(x):
    for subtype in ['H1N1','H1N2','H3N2']:
        if check_substring_in_file(x,subtype):
            return subtype
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Assuming df_coords is your DataFrame with 'lat', 'lon', 'emergence_score', and 'subtype' columns
subtypes = df_coords['subtype'].unique()
colors = plt.cm.cividis(np.linspace(0, 1, len(subtypes)))  # Generates a color map
colors = ['b','g','r']  # Generates a color map
subtype_color_map = dict(zip(subtypes, colors))

# Scaling the marker sizes (example using square of the emergence score)
marker_sizes = df_coords['irat_emergence_prediction'] ** 20
# Normalize the sizes for better visualization
marker_sizes = (marker_sizes - marker_sizes.min()) / (marker_sizes.max() - marker_sizes.min())
min_size, max_size = 3, 2000  # Minimum and maximum marker sizes
marker_sizes = marker_sizes * (max_size - min_size) + min_size

# Plotting
fig, ax = plt.subplots(1, 1, figsize=(15, 10))
world.plot(ax=ax, color='lightgrey', edgecolor='black')
for subtype in subtypes:
    subset = df_coords[df_coords['subtype'] == subtype]
    #plt.scatter(subset['lon'], subset['lat'], s=marker_sizes[subset.index], color=subtype_color_map[subtype], label=subtype)
    gdf = gpd.GeoDataFrame(
    subset, geometry=gpd.points_from_xy(subset.lon, subset.lat))
    gdf.plot(ax=ax, marker='o', color=subtype_color_map[subtype], alpha=.75, markersize=marker_sizes, label=subtype)
    
plt.title("World Map Showing Locations of Strains")

# Adding legend and other plot elements
plt.legend()
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Variants')



# Hide tick labels and axis labels
ax.set_xticklabels([])
ax.set_yticklabels([])
ax.set_xlabel('')
ax.set_ylabel('')

# Retain the grid
ax.grid(True)

# Adjust legend spacing and plot the legend
legend = ax.legend(title="Subtypes", bbox_to_anchor=(.9,.9), loc='upper left')
for handle in legend.legendHandles:
    handle.set_sizes([80])  # Adjust the size of the legend markers

plt.savefig('variants.png',dpi=600,bbpx_inches='tight',transparent=True)
```
