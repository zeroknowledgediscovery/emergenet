```python
import numpy as np
import pandas as pd
from quasinet.qnet import qdistance, load_qnet
from emergenet.domseq import DomSeq, save_model, load_model
import os
import shutil
```

```python
TAG0='A:American wigeon:South Carolina:AH0195145:2021_ha.fasta'
TAG=TAG0.replace(':','_').replace('.fasta','')
#TAG='A_American green-winged teal_Washington_1957050_2014_ha'
```

```python
# initialize the DomSeq
PATH='../../irat_qnet/raw_data/irat_sequences/'+TAG0
domseq = DomSeq(seq_trunc_length=550, random_state=42)
r0 = domseq.load_data(filepath=PATH)
r0=r0.set_index('id').sequence.values
r0_=np.array(list(r0[0]))
```

```python
Q_PATH='../../irat_qnet/qnet_models/'+TAG+'.joblib'
import zipfile
with zipfile.ZipFile(Q_PATH+'.zip', 'r') as zip_ref:
    zip_ref.extractall(Q_PATH)
if os.path.isdir(Q_PATH):
    shutil.move(Q_PATH,Q_PATH+'D')
    shutil.copy(Q_PATH+'D/'+Q_PATH.split('/')[-1],Q_PATH)
    
qnet=load_qnet(Q_PATH)

os.remove(Q_PATH)
shutil. rmtree(Q_PATH+'D')
```

```python
RAW_PATH='../../irat_qnet/raw_data/gisaid/'
```

```python
# initialize the DomSeq
domseq = DomSeq(seq_trunc_length=550, random_state=42)
df_ = domseq.load_data(filepath=RAW_PATH+TAG0)
```

```python
seq=df_.set_index('id').sequence.values
S=[np.array(list(x)) for x in seq]
```

```python
S_consensus=r0_
```

```python
import multiprocessing as mulpro
import itertools
```

```python
s0__=S_consensus
qnet__=qnet

def fpar(s):
    return qdistance(s0__,s,qnet__,qnet__)

def f(s_array):
    pool = mulpro.Pool(processes=10)
    return np.array(pool.map(fpar, s_array))    
```

```python
import shap
shap.initjs()
```

```python
explainer = shap.KernelExplainer(f, np.array([S_consensus]))
```

```python
S_=np.array([x for x in S if np.random.rand()<.15 ])
shap_values = explainer.shap_values(S_, nsamples=550)
```

```python
#shap.plots.bar(explainer(S__))
import pylab as plt
plt.style.use('dark_background')
shap.summary_plot(shap_values,S_, plot_type="bar",max_display=30)
```

```python
shp=pd.DataFrame(pd.DataFrame(shap_values).abs().mean().sort_values(ascending=False),columns=['shp'])
shp.index.name='features'
```
