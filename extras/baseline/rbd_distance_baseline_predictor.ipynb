```python
RBD=range(63,286)
#R=range(150,236)
#R=range(50,336)
#RBD=range(90,240)
binding_pocket=range(210,240)
df_h1n1=df#[df.subtype=='H3N2']
df_=df_h1n1[['emergence_risk','is_irat','ha_risk','geometric_mean_risk']]
R=[130,131,132,133,134,135,150,151,152,153,154,155,156,157,158,159,160,190,191,192,193,194,195,220,221,222,223,224,225,226]
df_['frag_rbd']=[''.join(np.array(list(x))[RBD]) for x in df_h1n1.ha.values]
df_['frag_sel']=[''.join(np.array(list(x))[R]) for x in df_h1n1.ha.values]
df_['frag_pocket']=[''.join(np.array(list(x))[binding_pocket]) for x in df_h1n1.ha.values]
df_['mean_geom_risk'] = df_.geometric_mean_risk.values
df_['mean_ha_risk'] = df_.ha_risk.values
#df_['max_ha_risk'] = [np.array(eval(x)).max() for x in df_.ha_risk]
#df_['min_ha_risk'] = [np.array(eval(x)).min() for x in df_.ha_risk]
#df_['max_geom_risk'] = [np.array(eval(x)).max() for x in df_.geometric_mean_risk]
#df_['min_geom_risk'] = [np.array(eval(x)).min() for x in df_.geometric_mean_risk]
df_=df_.drop(['ha_risk',
       'geometric_mean_risk'],axis=1)
```

```python
def getLav(s,seq0):
    from Levenshtein import distance
    return np.array([distance(s,x) for x in seq0]).min()
```

```python
high_risk_rbd = [getLav(x,seq0[:6]) for x in df_.frag_rbd.values]
df_['hr_rbd'] = high_risk_rbd
high_risk_rbd = [getLav(x,seq0[:6]) for x in df_.frag_sel.values]
df_['hr_sel'] = high_risk_rbd
high_risk_rbd = [getLav(x,seq0[:6]) for x in df_.frag_pocket.values]
df_['hr_pocket'] = high_risk_rbd
```

```python
import pandas as pd
import numpy as np
from scipy.stats import norm

def fisher_z_transform(r):
    return 0.5 * np.log((1 + r) / (1 - r))

def inverse_fisher_z(z):
    return (np.exp(2 * z) - 1) / (np.exp(2 * z) + 1)

def correlation_confidence_interval(data, alpha=0.05):
    corr = data.corr()
    n = data.shape[0]
    z = fisher_z_transform(corr)
    se = 1 / np.sqrt(n - 3)
    z_critical = norm.ppf(1 - alpha / 2)
    
    z_lower = z - z_critical * se
    z_upper = z + z_critical * se
    
    lower = inverse_fisher_z(z_lower)
    upper = inverse_fisher_z(z_upper)
    
    return corr, lower, upper

xf=df_[df_.is_irat==1].drop('is_irat',axis=1)
corr, lower, upper = correlation_confidence_interval(xf)
print("Correlation matrix:")
print(corr)
print("\nLower bounds of 95% confidence intervals:")
print(lower)
print("\nUpper bounds of 95% confidence intervals:")
print(upper)
```
