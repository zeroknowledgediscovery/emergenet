```python
def getLav(s,seq0):
    from Levenshtein import distance
    return np.array([distance(s,x) for x in seq0]).min()
```

```python
high_risk_rbd = [getLav(x,seq0) for x in df_.frag_rbd.values]
df_['hr_rbd'] = high_risk_rbd
high_risk_rbd = [getLav(x,seq0) for x in df_.frag_sel.values]
df_['hr_sel'] = high_risk_rbd
high_risk_rbd = [getLav(x,seq0) for x in df_.frag_pocket.values]
df_['hr_pocket'] = high_risk_rbd
```

```python
tf=df_[df_.is_irat==1].drop('is_irat',axis=1).corr().loc[['hr_rbd','hr_sel','hr_pocket'],:]
tf.index=['RBD','On seleted residues',"pocket"]
tf.index.name='edit distance'
tf.drop(['hr_rbd','hr_sel'],axis=1)
```
